```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import matplotlib.dates as mdates
import matplotlib
%matplotlib inline
```

```python
df = pd.read_csv('TSLA.csv')
```

```python
df.head(10)
```

```python
df['Total_Roll_Volume'] = df['Volume'].expanding(1).sum()
volume = df['Volume'].sum()
df['VWAP'] = df['Adj Close'] * df['Total_Roll_Volume'] / volume
```

```python
df.head(10)
```

```python
df.Date = pd.to_datetime(df.Date)
fig,ax = plt.subplots()
ax.plot_date(df.Date,df.Close)
ax.plot_date(df.Date,df.Open)
ax.plot_date(df.Date,df.VWAP)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%Y'))
plt.title('TSLA')
plt.xlabel('Date')
plt.ylabel('Price')
ax.legend()
fig.savefig('TSLA.pdf')
```
